```python
from manim import *
import numpy as np
```

```python
%%manim -ql FirstExample

class FirstExample(Scene):
    def construct(self):
        blue_circle = Circle(color=BLUE, fill_opacity=0.5)
        green_square = Square(color=GREEN, fill_opacity=0.8)
        green_square.next_to(blue_circle, RIGHT)
        self.add(blue_circle, green_square)
```

```python
%%manim -ql SecondExample

class SecondExample(Scene):
    def construct(self):
        ax = Axes(x_range=(-3, 3), y_range=(-3, 3))
        curve = ax.plot(lambda x: (x+2) * x * (x-2)/ 2, color=RED)
        area = ax.get_area(curve, x_range=range(-2, 0))
        self.play(Create(ax, run_time=2), Create(curve, run_time=5))
        self.play(FadeIn(area))
        self.wait(2)
```

```python
%%manim -ql SquareToCircle

class SquareToCircle(Scene):
    def construct(self):
        green_square = Square(color=GREEN, fill_opacity=0.5)
        self.play(DrawBorderThenFill(green_square))

        blue_circle = Circle(color=BLUE, fill_opacity=0.5)
        self.play(ReplacementTransform(green_square, blue_circle))
        self.play(FadeOut(blue_circle))
```

```python
%%manim -ql HelloManim

class HelloManim(Scene):
    def construct(self):
        text = Text("Hello, Manim!")
        self.play(Write(text))
        self.wait()
```

```python
%%manim -ql BasicShapes

class BasicShapes(Scene):
    def construct(self):
        circle = Circle()
        square = Square()
        triangle = Triangle()

        self.play(Create(circle))
        self.play(circle.animate.shift(LEFT))
        self.play(Transform(circle, square))
        self.play(Transform(circle, triangle))
        self.play(FadeOut(circle))
```

```python
%%manim -ql MathExpression

class MathExpression(Scene):
    def construct(self):
        equation = MathTex("E", "=", "mc^2")

        self.play(Write(equation))
        self.wait()
```

```python
%%manim -ql GradientDescent

class GradientDescent(Scene):
    def construct(self):
        #  
        def func(x):
            # return 0.2 * (x - 2) ** 2 + 1
            return x**4 - 5*x**2 - 3*x

        #   ()
        def dfunc(x):
            # return 0.4 * (x - 2)
            return 4 * x**3 - 10 * x - 3

        # Gradient Descent  
        def gradient_descent(x_start, learning_rate, epochs):
            points = [x_start]
            for _ in range(epochs):
                x_start -= learning_rate * dfunc(x_start)
                points.append(x_start)
            return points

        #  
        x_start = 0
        learning_rate = 0.2
        epochs = 10

        # x y 
        axes = Axes(x_range=[-3, 3, 1], y_range=[-15, 50, 10], x_length=10, y_length=6)

        #   
        # func_graph = FunctionGraph(func, x_range=[-2, 8], color=YELLOW)
        curve = axes.plot(func, color=YELLOW)

        self.play(Create(axes, run_time=2), Create(curve, run_time=3))

        # Gradient Descent  
        points = gradient_descent(x_start, learning_rate, epochs)

        #    
        for i in range(len(points) - 1):
            start = axes.c2p(points[i], func(points[i]))
            end = axes.c2p(points[i + 1], func(points[i + 1]))

            point = Dot(start, color=RED)
            self.play(FadeIn(point, run_time=0.5))

            arrow = Arrow(start, end, color=BLUE, buff=0)
            self.play(Create(arrow, run_time=0.5))

        #   
        last_point = Dot(axes.c2p(points[-1], func(points[-1])), color=RED)
        self.play(FadeIn(last_point, run_time=0.5))

        self.wait()
```
