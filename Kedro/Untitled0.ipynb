```python
from kedro.pipeline import Pipeline, node

def mean(xs, n):
    return sum(xs) / n


def mean_sos(xs, n):
    return sum(x*x for x in xs) / n


def variance(m, m2):
    return m2 - m * m


pipeline = Pipeline([
    node(func=len, inputs='xs', outputs='n'),
    node(func=mean, inputs=['xs', 'n'], outputs='m', name='mean_node'),
    node(func=mean_sos, inputs=['xs', 'n'], outputs='m2', name='mean_sos'),
    node(func=variance, inputs=['m', 'm2'], outputs='v', name='variance_node')
])
```

```python
print(pipeline.describe())
```
